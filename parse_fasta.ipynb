```python
def parse_fasta(filename):
    
    '''
    Parse fasta file into lists of sequence headers and sequences
    
    '''
    
    f = open(filename, 'r')
    data = f.readlines()
    f.close()

    sequences = []
    seq_names = []
    for line in data:
        if line[0] == '>':
            seq = ''
            sequences.append('')
            sn = line[1:-1]
            seq_names.append(sn)
        while line[0] != '>':
            sequences[-1] += line[0:-1]

            break
    
    return seq_names, sequences


```
